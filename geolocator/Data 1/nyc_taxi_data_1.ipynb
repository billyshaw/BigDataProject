```python
import pandas as pd
import os
```

```python
trip_data = pd.read_csv("trip_data_with_geo_1.csv")
trip_fare = pd.read_csv("trip_fare_1.csv")
```

Create unique trip_id for each row in both datasets for merging

```python
trip_data['trip_id'] = trip_data['medallion'] + '_' + trip_data['pickup_datetime']
trip_fare['trip_id'] = trip_fare['medallion'] + '_' + trip_fare['pickup_datetime']
```

Merge both data sets

```python
merged_data = trip_data.merge(trip_fare, on="trip_id")
```

```python
merged_data.to_csv("merged_data_1.csv")
```

Make column for fare percentage

```python
merged_data['fare_p'] = merged_data['tip_amount'] / merged_data['fare_amount']
```

Find basic information about taxi rides in current data sets

```python
merged_data.describe()
```

Find basic information about rides who pay with credit card (and, thus, tips)

```python
data_descr = merged_data.query("payment_type == 'CRD'").describe()
```

```python
neighborhood_descr = merged_data.groupby("pickup_neighborhood").describe()
```

```python
data_descr
neighborhood_descr
```

Find basic information about rides by hour

```python
hour_descr = merged_data.groupby("pickup_hour").describe()
```

Find count of rides per pickup neighborhood

```python
neighborhood_count = merged_data.groupby("pickup_neighborhood").count()
```

```python
neighborhood_count['count_p'] = neighborhood_count['fare_p'] / neighborhood_count['fare_p'].sum()
```

```python
neighborhood_count['count_p'].order(ascending=False)[0:19]
```

Find count of rides per dropoff neighborhood

```python
neighborhood_dcount = merged_data.groupby("dropoff_neighborhood").count()
neighborhood_dcount['count_p'] = neighborhood_count['fare_p'] / neighborhood_count['fare_p'].sum()
```

```python
neighborhood_dcount['count_p'].order(ascending=False)[0:19]
```

Find count of rides per time of pickup

```python
time_count = merged_data.groupby("pickup_hour").count()
```

```python
time_count['count_p'] = time_count['fare_p'] / time_count['fare_p'].sum()
time_count['count_p'].order(ascending=False)[0:19]
```

Find count of rides per time of dropoff

```python
time_dcount = merged_data.groupby("dropoff_hour").count()
time_dcount['count_p'] = time_dcount['fare_p'] / time_dcount['fare_p'].sum()
time_dcount['count_p'].order(ascending=False)[0:19]
```

Find time w/ the highest average speed

```python
time = merged_data.groupby("pickup_hour").mean()
time['speed'] = time['trip_distance'] / time['trip_time_in_secs']
```

```python
time['speed'].order(ascending=False)
```
