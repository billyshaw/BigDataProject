```python
import pandas as pd
import os
```

Import data

```python
trip_data = pd.read_csv("trip_data_with_geo_1.csv")
```

```python
trip_fare = pd.read_csv("trip_fare_1.csv")
```

Rename the columns and remove the space

```python
trip_fare = trip_fare.rename(columns = {' hack_license':'hack_license', ' vendor_id':'vendor_id', ' pickup_datetime':'pickup_datetime', \
                                                                ' payment_type':'payment_type', ' fare_amount':'fare_amount',' surcharge':'surcharge', ' mta_tax':'mta_tax',\
                                                                ' tip_amount':'tip_amount',' tolls_amount':'tolls_amount',' total_amount':'total_amount' } )
```

Create unique trip_id for each row in both datasets for merging

```python
trip_data['trip_id'] = trip_data['medallion'] + '_' + trip_data['pickup_datetime']
```

```python
trip_fare['trip_id'] = trip_fare['medallion'] + '_' + trip_fare['pickup_datetime']
```

Merge both data sets and make a column for tip as a percentage of cab fare


```python
merged_data = trip_data.merge(trip_fare, on="trip_id")
```

```python
merged_data['fare_p'] = merged_data['tip_amount'] / merged_data['fare_amount']
```

```python
merged_data.to_csv("merged_data_1.csv")
```

Find count of rides per pickup neighborhood

```python
neighborhood_count = merged_data.groupby("pickup_neighborhood").count()
neighborhood_count['count_p'] = neighborhood_count['fare_p'] / neighborhood_count['fare_p'].sum()
top_neighborhood = neighborhood_count['count_p'].order(ascending=False)
```

```python
top_neighborhood.to_csv("top_neighborhood.csv")
```

Find count of rides per dropoff neighborhood

```python
neighborhood_dcount = merged_data.groupby("dropoff_neighborhood").count()
neighborhood_dcount['count_p'] = neighborhood_count['fare_p'] / neighborhood_count['fare_p'].sum()
```

```python
top_drop_neighborhood = neighborhood_dcount['count_p'].order(ascending=False)
```

```python
top_drop_neighborhood.to_csv("top_drop_neighborhood.csv")
```

Find count of rides per time of pickup

```python
time_count = merged_data.groupby("pickup_hour").count()
```

```python
time_count['count_p'] = time_count['fare_p'] / time_count['fare_p'].sum()
count_by_hour = time_count['count_p'].order(ascending=False)
```

```python
count_by_hour
```

```python
count_by_hour.to_csv("count_by_hour.csv")
```

Find time w/ the highest average speed

```python
time = merged_data.groupby("pickup_hour").mean()
time['speed'] = time['trip_distance'] * 3600 / time['trip_time_in_secs']
time_speed = time['speed'].order(ascending=False)
```

```python
time_speed.to_csv("time_speed.csv")
```
